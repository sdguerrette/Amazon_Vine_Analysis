```python
import pandas as pd
```

```python
# import vine_table csv into pandas as dataframe
df = pd.read_csv("vine_table.csv")
df
```

```python
# create new dataframe with only rows where total_vote is 20 or more
helpful_df = df[df['total_votes']>=20]
helpful_df
```

```python
# create new dataframe where the percentage of helpful votes to total votes is greater than .5
filtered_df = helpful_df[(helpful_df['helpful_votes']/helpful_df['total_votes'])>=.5]
filtered_df
```

```python
# create new dataframe with only vine reviews
vine_df = filtered_df[filtered_df['vine']=='Y']
vine_df
```

```python
# create new dataframe with only non-vine (unpaid) reviews
no_vine_df = filtered_df[filtered_df['vine']=='N']
no_vine_df
```

```python
# calculate the number of vine reviews
paid_count = vine_df['review_id'].count()
paid_count
```

```python
# calculate the number of 5 star vine reviews
five_star_paid = vine_df[vine_df['star_rating']==5].count()['review_id']
five_star_paid
```

```python
#calculate the percent of 5 star vine reviews
paid_five_star_percent = five_star_paid/paid_count*100
paid_five_star_percent
```

```python
# calculate the number of unpaid reviews
unpaid_count = no_vine_df['review_id'].count()
unpaid_count
```

```python
# calculate the numner of 5 star unpaid reviews
five_star_unpaid = no_vine_df[no_vine_df['star_rating']==5].count()['review_id']
five_star_unpaid
```

```python
# calculate the percent of 5 star unpaid reviews
unpaid_five_star_percent = five_star_unpaid/unpaid_count*100
unpaid_five_star_percent
```
